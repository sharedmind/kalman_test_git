```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
import numpy as np
```

```python
output_notebook()
```

```python
def lms_adapt(N = 21, Ns = 100, beta = 0.01):
    """
    x - input data buffer to the filter
    N - filter order
    beta - learning rate
    Ns - number of samples
    y_out - filter output
    """
    
    y_out = np.zeros(Ns)
    desired = np.zeros(Ns)
    error = np.zeros(Ns)
    x = np.zeros(N)#bufor wejscie o dlugosci okna N
    w = np.zeros(N)#wagi-wspolczynniki filtru
    for t in range(0,Ns):
        x[0] = np.sin(2.0*np.pi*t/8.0)#pobierz nowa probke wejsciowa
        d = 2*np.cos(2*np.pi*t/8.0)#pobierz nowa probke sygnalu pozadanego
        y = 0
        for i in range(0,N):
            y+=w[i]*x[i]#oblicz wyjscie filtru
        e = d - y#oblicz blad
        for i in range(N-1,-1,-1):
            w[i]+=beta*e*x[i]
            if i!=0:
                x[i]=x[i-1]
        desired[t] = d
        y_out[t] = y
        error[t] = e
    return error, desired, y_out

t = np.linspace(0,10,100)
e,d,y = lms_adapt()
p1 = figure(plot_width = 650, plot_height=350)
p1.line(t,y, legend="y_out", color="red", line_width=2)
p1.line(t,d, legend="desired", color="blue", line_width=2)
show(p1)
        
    
```
